```python
# Definisi graf berbobot dalam bentuk dictionary of dictionaries.
# Setiap node memiliki tetangga dan bobot (biaya) menuju tetangga tersebut.
graph = {
    'A': {'B': 2, 'C': 3},
    'B': {'D': 4},
    'C': {'D': 1, 'F': 2},
    'D': {'E': 5},
    'E': {'F': 2},
    'F': {'G': 3},
    'G': {}
}

def hill_climb_path(graph, start, goal):
    # Fungsi untuk mencari jalur dari start ke goal dengan pendekatan Hill Climbing.
    # Pilihan ditentukan berdasarkan tetangga dengan biaya terendah (greedy).
    path = [start]
    current = start
    visited = set()
    
    while current != goal:
        # TUGAS ANDA (Nilai 20)
        # Loop berjalan hingga node tujuan tercapai atau tidak ada langkah lagi.
        # Ambil semua tetangga dari node saat ini.
        # Inisialisasi nilai minimum dan node berikutnya yang akan dikunjungi.
        neighbors = graph[current]
        min_cost = float('inf')
        next_node = None
        
        # TUGAS ANDA (Nilai 40)
        # Pilih tetangga dengan biaya terkecil yang belum dikunjungi.
        for n, cost in neighbors.items():
            if n not in visited and cost < min_cost:
                min_cost = cost
                next_node = n
        
        # TUGAS ANDA (Nilai 20)
        # Jika tidak ada tetangga yang bisa dikunjungi, hentikan proses.
        # Tambahkan node tersebut ke dalam path.
        # Tandai node saat ini sebagai sudah dikunjungi.
        # Pindah ke node berikutnya.
        if next_node is None:
            break
        
        path.append(next_node)
        visited.add(current)
        current = next_node

    return path

# TUGAS ANDA: (Nilai 20)
# Cetak jalur hasil pencarian dari node 'A' ke 'G', dipisahkan dengan panah.
result_path = hill_climb_path(graph, 'A', 'G')
print(" -> ".join(result_path))
```

```python
import random, math  # Import pustaka random dan math

def distance(a, b):
    # Tugas Anda: Nilai 20
    # Jarak Euclidean
    return math.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

def evaluate(sensors):
    overlap = 0
    min_distance = 2.0
    for i in range(len(sensors)):
        for j in range(i+1, len(sensors)):
            # Tugas Anda: Nilai 20
            # Hitung pasangan sensor terlalu dekat
            dist = distance(sensors[i], sensors[j])
            if dist < min_distance:
                overlap += 1
    return overlap

def simulated_annealing_sensors(T=100, alpha=0.95):
    sensors = [(random.uniform(0, 10), random.uniform(0, 10)) for _ in range(5)]  # Inisialisasi
    current_score = evaluate(sensors)
    while T > 0.01 and current_score > 0:
        new_sensors = sensors[:]
        i = random.randint(0, 4)
        new_sensors[i] = (random.uniform(0, 10), random.uniform(0, 10))
        # Tugas Anda: Nilai 40
        # Ubah posisi satu sensor
        # Evaluasi solusi baru
        new_score = evaluate(new_sensors)
        delta = new_score - current_score
        
        if delta < 0 or math.exp(-delta / T) > random.random():
            sensors = new_sensors
            current_score = new_score
            
        T *= alpha  # Pendinginan
    return sensors, current_score

# Tugas Anda: Nilai 20
# Jalankan algoritma
# Cetak hasil
for run in range(2):
    print(f"\n--- RUN {run + 1} ---")
    sensors, score = simulated_annealing_sensors(T=100, alpha=0.95)
    
    print(f"\nHasil Run {run + 1}:")
    print(f"Score akhir (overlap): {score}")
    print("Posisi sensor:")
    for i, sensor in enumerate(sensors):
        print(f"  Sensor {i+1}: ({sensor[0]:.3f}, {sensor[1]:.3f})")
```
